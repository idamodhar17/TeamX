```python
import ollama

messages = [
    {
        'role': 'health cosaltant',
        'content': 'you have a health consultant and you need to consult the tretment for the patient depending on their disease and also give healpline number like ambulence etc. in ind and in case of emergency call on emergency helpline no.'
    }
]

def truncate_response(response_text, max_words=100):
    words = response_text.split()
    if len(words) > max_words:
        return ' '.join(words[:max_words]) + '...'
    return response_text
while True:
   
    user_input = input("You: ")

    messages.append({'role': 'user', 'content': user_input})

    response = ollama.chat(model='llama3.1', messages=messages)
    
    ai_response = response['message']['content']
    truncated_response = truncate_response(ai_response, max_words=100)
    print(f"AI: {truncated_response}")
    
    messages.append({'role': 'assistant', 'content': truncated_response})
```
